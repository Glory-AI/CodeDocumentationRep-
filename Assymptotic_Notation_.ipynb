 Think of asymptotic notation like a set of instructions where you're willing to vary the input size â€” but still want your program to remain efficient and not take forever to run. Thatâ€™s where asymptotic notation comes in.

Itâ€™s a useful concept that models the relationship between input size and algorithm running time.

Most of the time, itâ€™s time complexity we notice, while space complexity only becomes visible when memory runs out ðŸŒš. Thatâ€™s why you need to be familiar with this and know how to apply it.

Big O notation is the most commonly used.
It gives the upper bound â€” meaning how bad the running time of a function or algorithm can get.

Some examples:

ðŸ”¹ O(1) â€“ Constant time (running time doesn't grow with input)

ðŸ”¹ O(n) â€“ Linear time (grows with input size)

ðŸ”¹ O(nÂ²) â€“ Quadratic time (nested loops)

ðŸ”¹ O(log n) â€“ Logarithmic time (cuts input down in steps)

ðŸ”¹ O(n log n) â€“ Log-linear time (efficient sorting and searching)

Itâ€™s all about writing efficient programs.
Sometimes it wonâ€™t be the obvious solution â€” there might be a few tricky ideas to understand ðŸ˜…, but once you get it, youâ€™ve unlocked something powerful.

```python
#an updated version of the learn_course function
def learn_course(topic, subtopics_dict):
    response = input(f"Do you understand {topic}? Type 'Understood' or 'No': ")
    if response == 'Understood':
        return f"Move on from {topic}"
    else:
        print(f"Break down {topic} and seek help to understand...")

        # Go through subtopics
        subtopics = subtopics_dict.get(topic, [])
        for sub in subtopics:
            # Check if sub has sub-subtopics using isinstance which checks for
            # datatype and the kind we need here is a group(tuple)
            #If sub is a tuple: It means this subtopic has a list of sub-subtopics associated with it. The code then extracts the subtopic name and its sub-subtopics list and calls learn_course recursively to process the sub-subtopics.
            #If sub is not a tuple: It means it's a simple subtopic without further nesting. The code calls learn_course on this subtopic with an empty dictionary for subtopics, indicating that it has no further breakdown
            if isinstance(sub, tuple):
                subtopic, subsub_list = sub #tuple unpacking
                learn_course(subtopic, {subtopic: subsub_list})
            else:
                learn_course(sub, {})

        return f"Now, topic '{topic}' is fully understood, move on to the next."

# Example usage
topic = "EEE254"
subtopics_structure = {
    "EEE254": [
        ("Scoping", ["Local Scope", "Global Scope", "Built-in Scope"]),
        ("Stacks", ["Function Call Stack", "Stack Overflow"]),
        "Recursion"
    ]
}

learn_course(topic, subtopics_structure)



#O(n^2) - Quadratic time (complexity growing as twice as size of input)
#if each subtopic has their own sub-subtopics
#  recursive calls inside loops, again and again.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample input sizes (n values)
#linspace function is responsible for generating the sequence of numbers n
n = np.linspace(1, 50, 100)

# Big O functions
O_1 = np.ones_like(n)  #creates a new size of n filled with ones(i.e regardless
#of input size, running time remains the same)    # Constant
O_log_n = np.log2(n)                            # Logarithmic
O_n = n                                         # Linear
O_n_log_n = n * np.log2(n)                      # Log-linear
O_n_squared = n**2                              # Quadratic


# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, O_1, label="O(1) - Constant", linewidth=2)
plt.plot(n, O_log_n, label="O(log n) - Logarithmic", linewidth=2)
plt.plot(n, O_n, label="O(n) - Linear",  linewidth=2)
plt.plot(n, O_n_log_n, label="O(n log n) - Log-Linear" , linewidth=2)
plt.plot(n, O_n_squared, label="O(nÂ²) - Quadratic", linewidth= 2)

# Customize the plot
plt.title("Big O Time Complexities", fontsize=14)
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)

# Show the plot
plt.show()

```
