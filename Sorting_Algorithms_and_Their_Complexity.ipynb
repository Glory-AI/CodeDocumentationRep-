There are different ways to sort things in programming, and they’re called sorting algorithms.

1. Bubble sort is like checking one item against the next over and over. If one is bigger, you swap them. Keep doing that till the whole list is sorted.

2. Selection sort works by finding the smallest item in the list and putting it first, then finding the next smallest and so on.

3. Insertion sort builds the sorted list one item at a time. It picks an item and slides it into the right spot among the ones already sorted.

4. Quick sort is smart and fast. It picks a pivot value, splits the list into two — smaller values on the left, bigger on the right — and sorts each side.

5. Merge sort breaks the list into smaller chunks, sorts them, and merges everything back in order.

```python
import random
import time
import matplotlib.pyplot as plt

# Generate 1000 random voltage-current readings into a dictionary
voltage_current_data = {round(random.uniform(0, 100), 2): round(random.uniform(0, 20), 2) for _ in range(1000)} # use _ when you don’t need to reference the loop counter in the loop body

# Convert to list of tuples for sorting
readings = list(voltage_current_data.items())
#Beefore searching, it is advisable to sort data
#So the following functions are implementations of sorting algorithms.

# -------- BUBBLE SORT --------
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j][0] > arr[j + 1][0]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# -------- SELECTION SORT --------
def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j][0] < arr[min_index][0]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# -------- INSERTION SORT --------
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key[0] < arr[j][0]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# -------- MERGE SORT --------
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i][0] < R[j][0]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

# -------- QUICK SORT --------
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2][0]
    left = [x for x in arr if x[0] < pivot]
    middle = [x for x in arr if x[0] == pivot]
    right = [x for x in arr if x[0] > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# -------- USING THE QUICK SORT FOR VOLTAGE_CURRENT EXPERIMENT TRACKER--------
sorted_data = quick_sort(readings.copy())
voltages = [v[0] for v in sorted_data]
currents = [v[1] for v in sorted_data]

plt.plot(voltages, currents)
plt.title('Voltage vs Current (Sorted by Voltage)')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.grid(True)
plt.show()
```
