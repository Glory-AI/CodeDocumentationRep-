```python
from collections import Counter
import matplotlib.pyplot as plt
import csv
import os
#mood_tracker
#taking care of all days in a week
moods= []
days= ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
for m in range(7):
  mood= str(input(f"How did you feel on {days[m]} (e.g happy,anxious,stressed)? ")).strip().lower()
  moods.append(mood)
print('Your mood throughout the week(Monday-Thursday):', moods)


mood_counts = Counter(moods)

# Display summary
print("\nMood summary:")
#accessing dictionary to display mood(key) and count(value)
for mood, count in mood_counts.items():
    print(f"You were {mood.capitalize()} {count} time(s) this week")
    if count >=3:
      print(f"ðŸ‘‰ You were really {mood} this week!")
#plotting pie chart to observe distribution
plt.pie(mood_counts.values(), labels=mood_counts.keys(), autopct='%1.1f%%')
plt.title('Mood Distribution Of The Week')
plt.show()



```

```python
# Setting up file
filename = "moods_per_week.csv"
file_exists = os.path.isfile(filename)

# Determine current week number in file mode
if not file_exists:
    week_number = 1
else:
    with open(filename, mode='r') as file:
        lines = file.readlines()
        week_numbers = []
        for line in lines[1:]:  # Skip header
            try:
                week = int(line.split(',')[0])
                week_numbers.append(week)
            except:
                continue
        week_number = max(week_numbers) + 1 if week_numbers else 1

```

```python
# Save to CSV
with open(filename, mode='a', newline='') as file:
    writer = csv.writer(file)
    if not file_exists:
        writer.writerow(["Week", "Day", "Mood"])  # write header only once
    for day, mood in zip(days, moods):
      #for loop unpacks each tuple (pair) created by zip() and writes into fi
        writer.writerow([week_number, day, mood])

print(f"\nâœ… Mood data for Week {week_number} saved successfully to '{filename}'.")
```
